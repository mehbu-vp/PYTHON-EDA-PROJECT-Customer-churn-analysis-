```python
#CLIENTNUM: Unique identifier for each client or customer.
Attrition_Flag: Binary variable indicating whether the customer has churned (attrition) or not.
Customer_Age: Age of the customer.
Gender: Gender of the customer.
Dependent_count: Number of dependents of the customer.
Education_Level: Education level of the customer.
Marital_Status: Marital status of the customer.
Income_Category: Income category of the customer.
Card_Category: Category of the credit card the customer holds.
Months_on_book: Number of months the customer has been a client.
Total_Relationship_Count: Total number of products held by the customer.
Months_Inactive_12_mon: Number of months with no transaction activity in the last 12 months.
Contacts_Count_12_mon: Number of contacts or interactions with the customer in the last 12 months.
Credit_Limit: Credit limit for the customer.
Total_Revolving_Bal: Total revolving balance on the credit card.
Avg_Open_To_Buy: Average open-to-buy credit line (Credit_Limit - Total_Revolving_Bal).
Total_Trans_Amt: Total transaction amount in the last 12 months.
Total_Trans_Ct: Total number of transactions in the last 12 months.
Avg_Utilization_Ratio: Average utilization ratio of the credit card.
```

# DATA CLEANING

```python
#check null values 
customer_data.isnull().sum()
```

```python
#null value treatment(imputation)
customer_data['Gender'].fillna(customer_data['Gender'].mode()[0],inplace=True)
customer_data['Card_Category'].fillna(customer_data['Card_Category'].mode()[0],inplace=True)
customer_data['Credit_Limit'].fillna(customer_data['Credit_Limit'].mode()[0],inplace=True)
```

```python
#check duplicates
customer_data.duplicated()
```

```python
#identify the categorical and numerical data
cat=[]
num=[]
def data_type(df):
    for col in df.columns:
        if df[col].nunique()<30:
            cat.append(col)
        else:
            num.append(col)
            
data_type(customer_data)
```

```python
cat
```

```python
num
```

```python
def mode_cat(cat):
    for col in cat:
        print(f'{col},Mode={customer_data[col].mode()[0]},Nunique {customer_data[col].nunique()} , Missing_Val {customer_data[col].isnull().sum()}')
        
mode_cat(cat)    
```

# EDA

```python
#Univariate Analysis
```

```python
# Distribution of Customer Age
plt.figure(figsize=(10, 6))
sns.histplot(customer_data['Customer_Age'], bins=20, kde=True)
plt.title('Distribution of Customer Age')
plt.show()
```

```python

# Distribution of Customer Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=customer_data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Count of Attrition_Flag
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition_Flag', data=customer_data)
plt.title('Count of Attrition_Flag')
plt.show()
```

```python
# Income Category Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Income_Category', data=customer_data)
plt.xticks(rotation=45)
plt.title('Income Category Distribution')
plt.show()
```

```python
#Bivariate Analysis
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Income_Category', y='Credit_Limit', data=customer_data)
plt.title('Credit Limit by Income Category')
plt.xlabel('Income Category')
plt.ylabel('Credit Limit')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship between Age and Total Transaction Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Customer_Age', y='Total_Trans_Amt', data=customer_data, hue='Attrition_Flag')
plt.title('Age vs Total Transaction Amount')
plt.show()
```
